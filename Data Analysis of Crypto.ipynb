```python
#python code for data visualisation of cryptocurrencies
import pandas as pd
import numpy as np


#reading datasets
btc = pd.read_csv('btc_usd.csv')
doge = pd.read_csv('doge_usd.csv')
eth = pd.read_csv('eth_usd.csv')
shib = pd.read_csv('shib_usd.csv')
aave = pd.read_csv('aave_usd.csv')
link = pd.read_csv('link_usd.csv')
```

```python
#Creating a new dataframe that holds the closing price of all 4 crypto currencies
dataf=pd.DataFrame({'BTC':btc['Close'],
                     'DOGE':doge['Close'],
                     'ETH':eth['Close'],
                     'SHIB':shib['Close'],
                     'AAVE':aave['Close'],
                     'LINK':link['Close']
                   })
```

```python
#The Dataframe dataf will look like
dataf
```

```python
#Analsis of Data
dataf.describe()
```

```python
#Visualizing the closing prices 
#Importing matplotlib.pyplot 

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

my_frame=dataf
plt.figure(figsize=(12.2,7.5))
for coin in my_frame.columns.values:
    plt.plot(my_frame[coin], label=c)
    
plt.title('Closing prices')
plt.xlabel('Days')
plt.ylabel('Price in USD')
plt.legend(my_frame.columns.values,loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

my_frame=dataf

exc_btc = dataf[dataf.columns.difference(['BTC'])]

plt.figure(figsize=(12.2,7.5))
for c in exc_btc.columns.values:
    plt.plot(exc_btc[c], label=c)
    
plt.title('Closing prices excluding BTC')
plt.xlabel('Days')
plt.ylabel('Price in USD')
plt.legend(exc_btc.columns.values,loc='upper left')
plt.show()
```

```python
#Scaling the data
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,100))
scaled = min_max_scaler.fit_transform(dataf)
scaled
```

```python
#Converting the scaled data into the dataframe
dataf_scaled = pd.DataFrame(scaled,columns=dataf.columns)
#visualize the scaled data
my_crypto = dataf_scaled

plt.figure(figsize = (13.5,5.2))
for c in my_crypto.columns.values:
    plt.plot(my_crypto[c],label = c)


plt.title('Cryptocurrency Scaled Graph')
plt.xlabel('Days')
plt.ylabel('Crypto Scaled Price in USD')
plt.legend(my_crypto.columns.values, loc = 'upper left')
plt.show()
```

```python
#Getting the daily simple return
DSR = dataf.pct_change(1)
DSR
```

```python
#Visuvalizing the daily simple returns
plt.figure(figsize=(19.5,8.5))
for c in DSR.columns:
    plt.plot(DSR.index,DSR[c], label=c,lw=2,alpha=.7)
    
plt.title('Daily simple returns')
plt.ylabel('Percentage in decimal form ')
plt.xlabel('Days')
plt.legend(DSR.columns.values,loc='upper right')
plt.show()
```

```python
#standard deviation is a measure of volatility of a coin
print('The cryptocurrency volatility')
np.std(DSR)
```

```python
#mean of coin
np.mean(DSR)
```

```python
#correlation
DSR.corr()
```

```python
#Getting the daily cumulative simple returns
DCSR=(DSR+1).cumprod()

#print DSR
DCSR
```

```python
#visualizing the daily cumulative simple returns
plt.figure(figsize=(12.3,5))
for c in DCSR.columns.values:
    plt.plot(DCSR.index, DCSR[c] ,lw=2,label=c)
    
plt.title('Daily Cumulative simple returns')
plt.xlabel('Days')
plt.ylabel('Growth of $1 investment')
plt.legend(DCSR.columns.values, loc='upper left' ,fontsize=18)
plt.show()
```

```python
#visualizing the daily cumulative simple returns

newdf = DCSR[DCSR.columns.difference(['DOGE'])]

plt.figure(figsize=(12.3,5))
for c in newdf.columns.values :
    plt.plot(newdf.index, newdf[c] ,lw=2,label=c)
    
plt.title('Daily Cumulative simple returns (excluding DOGE)')
plt.xlabel('Days')
plt.ylabel('Growth of $1 investment')
plt.legend(newdf.columns.values, loc='upper left' ,fontsize=18)
plt.show()


```
